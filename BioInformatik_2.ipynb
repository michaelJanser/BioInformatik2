## Das modulare Ökosystem-Labor: Regulierende Faktoren

Willkommen zurück im Software-Testlabor! In der ersten Phase habt ihr erfolgreich das chaotische, kartenbasierte Grundmodell gezähmt. Ihr habt an den biologischen Stellschrauben gedreht und eine stabile, zeitversetzte **Lotka-Volterra-Schwingung** erzeugt.

Heute verknüpfen wir diese mathematische Ideallösung mit der biologischen Realität. In der echten Natur schwingen Populationen nicht isoliert im luftleeren Raum. Sie werden permanent von der Umwelt beeinflusst.

Wir unterscheiden dabei zwei fundamentale Arten von Einflüssen:
1. **Dichteunabhängige Faktoren:** Einflüsse (meistens vom Wetter oder Klima), die das Ökosystem wie ein Blitz aus heiterem Himmel treffen — völlig egal, wie viele Tiere gerade im Wald leben.
2. **Dichteabhängige Faktoren:** Einflüsse, die umso stärker wirken, je dichter und voller eine Population wird (z.B. Nahrungsmangel, ansteckende Krankheiten oder Stress).

## Eure Programmier-Challenge 🏆

Jede Gruppe wählt nun aus dem folgenden Rezeptbuch **einen dichteunabhängigen** und **einen dichteabhängigen** Faktor aus.

Scrollt in der nächsten Code-Zelle zu **SLOT A** und **SLOT B**. Dort ist das Gerüst für eure gewählte Option bereits vorbereitet. Euer Auftrag ist es, die Platzhalter (wie `1.0` oder `0`) so anzupassen, dass die biologischen Regeln korrekt ausgeführt werden!

---

### 🔵 BOX 1: Dichteunabhängige Faktoren (Wählt eine Option für SLOT A)

* **Option A1: Der Jahrhundertwinter**
    * **Die Biologie:** In jeder Runde besteht eine Chance von genau 4% (`0.04`), dass ein harter Winter einbricht.
    * **Euer Code-Auftrag:** Wenn der Winter zuschlägt, sterben schlagartig 30% der Mäuse und 20% der Füchse.
    * *Mathematischer Tipp:* Wenn 30% sterben, überleben 70%. Ihr müsst die Population also mit `0.7` multiplizieren. Berechnet den Faktor für die Füchse analog!

* **Option A2: Die Sommer-Flut**
    * **Die Biologie:** In jeder Runde besteht eine Chance von genau 3% (`0.03`), dass ein Hochwasser die Mäusebauten überschwemmt. Nur 30% der Mäuse überleben. Die Füchse sind nicht betroffen.
    * **Euer Code-Auftrag:** Passt die Wahrscheinlichkeit bei der `if`-Abfrage an und reduziert die Mäuse auf die überlebenden 30%.

---

### 🟢 BOX 2: Dichteabhängige Faktoren (Wählt eine Option für SLOT B)

* **Option B1: Die Mäuse-Seuche**
    * **Die Biologie:** Wenn zu viele Mäuse auf engem Raum leben, breiten sich Krankheiten rasant aus.
    * **Euer Code-Auftrag:** Passt den Schwellenwert in der `if`-Bedingung auf **180 Mäuse** an. Wenn dieser erreicht wird, soll der Code die Mäusepopulation schlagartig halbieren (Multiplikation mit `0.5`).

* **Option B2: Die Greifvogel-Invasion**
    * **Die Biologie:** Wenn es im Tal von Mäusen wimmelt, lockt das Raubvögel an. Sie schnappen den Füchsen die Nahrung weg.
    * **Euer Code-Auftrag:** Passt den Schwellenwert auf **130 Mäuse** an. Wenn dieser überschritten wird, müssen den Füchsen jede Runde zusätzlich **15 Mäuse** abgezogen werden.

* **Option B3: Der Nahrungsmangel (Das Gras-Experiment) 🌾**
    * **Die Biologie:** Mäuse fressen Gras. Im Hintergrund wächst jede Runde Gras nach. Wenn zu viele Mäuse alles kahlfressen, setzt eine Hungersnot ein.
    * **Euer Code-Auftrag:** Das Gerüst prüft bereits, ob das Gras knapp wird (`if gras < maeuse:`). Wenn ja, wird die Mäuseanzahl auf den aktuellen Grasbestand gesetzt. Ihr müsst nun ganz unten die Zeile schreiben, die das gefressene Gras vom Gesamten abzieht: `gras = gras - ...`

```python
import random
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Startwerte & Parameter
maeuse = 3
fuechse = 1
runden = 200
gras = 400

# Listen für die Grafik-Historie
historie_maeuse = [maeuse]
historie_fuechse = [fuechse]

for runde in range(1, runden + 1):

    # Im Hintergrund: Gras wächst jede Runde nach (wichtig für Option B3)
    gras = min(600, gras + 120)

    # =================================================================
    # 📑 SLOT A: DICHTEUNABHÄNGIGER FAKTOR (Wetter)
    # AUFGABE: Passt die Zahlen bei EURER gewählten Option an!
    # Die andere Option ignoriert ihr einfach.
    # =================================================================

    # --- [OPTION A1: Jahrhundertwinter] ---
    if random.random() < 0.00: # <-- TODO: Wahrscheinlichkeit auf 0.04 anpassen
        print(f"❄️ Runde {runde}: Jahrhundertwinter!")
        maeuse = int(maeuse * 1.0)   # <-- TODO: Faktor für 30% Verlust anpassen
        fuechse = int(fuechse * 1.0) # <-- TODO: Faktor für 20% Verlust anpassen

    # --- [OPTION A2: Sommer-Flut] ---
    if random.random() < 0.00: # <-- TODO: Wahrscheinlichkeit auf 0.03 anpassen
        print(f"🌊 Runde {runde}: Sommer-Flut!")
        maeuse = int(maeuse * 1.0)   # <-- TODO: Faktor für 70% Verlust anpassen

    # =================================================================

    # --- REVIERKONKURRENZ & ABWANDERUNG (Füchse) ---
    aktive_jagd_fuechse = 0
    for f in range(fuechse):
        ueberlappungs_chance = aktive_jagd_fuechse / 32.0
        if random.random() >= ueberlappungs_chance:
            aktive_jagd_fuechse += 1

    # --- DIE REALE JAGD ---
    insgesamt_gefressen = 0
    fuechse_naechste_runde = 0
    maus_dichte = maeuse / 234.0
    if maus_dichte > 1.0:
        maus_dichte = 1.0

    for f in range(aktive_jagd_fuechse):
        fang = int(6 * maus_dichte)
        if insgesamt_gefressen + fang > maeuse:
            fang = maeuse - insgesamt_gefressen
        insgesamt_gefressen += fang

        # Fuchs-Biologie
        if fang < 2: pass
        elif fang == 2: fuechse_naechste_runde += 1
        elif 3 <= fang <= 4: fuechse_naechste_runde += 2
        elif fang >= 5: fuechse_naechste_runde += 3

    # --- BIOLOGIE MAUS ---
    ueberlebende_maeuse = maeuse - insgesamt_gefressen
    if ueberlebende_maeuse > 0:
        maeuse = int(ueberlebende_maeuse * 1.4)
    else:
        maeuse = 0

    # =================================================================
    # 📑 SLOT B: DICHTEABHÄNGIGER FAKTOR (Regulation)
    # AUFGABE: Passt die Zahlen bei EURER gewählten Option an!
    # =================================================================

    # --- [OPTION B1: Mäuse-Seuche] ---
    if maeuse > 1000: # <-- TODO: Schwellenwert auf 180 anpassen
        print(f"🦠 Runde {runde}: Eine Seuche bricht aus!")
        maeuse = int(maeuse * 1.0) # <-- TODO: Population halbieren (Faktor anpassen)

    # --- [OPTION B2: Greifvogel-Invasion] ---
    if maeuse > 1000: # <-- TODO: Schwellenwert auf 130 anpassen
        print(f"🦅 Runde {runde}: Greifvögel jagen mit!")
        maeuse = maeuse - 0 # <-- TODO: 15 Mäuse abziehen

    # --- [OPTION B3: Gras-Experiment] ---
    if gras < maeuse:
        print(f"🌾 Runde {runde}: Hungersnot! Wiese kahlgefressen.")
        maeuse = gras
    # <-- TODO: Schreibt hier drunter die Zeile, die das gefressene Gras abzieht!
    # Tipp: gras = gras - ...

    # =================================================================

    # --- REGEL-LIMITS & NOT-PATCHES ---
    fuechse = fuechse_naechste_runde
    if maeuse > 234: maeuse = 234
    if fuechse > 36: fuechse = 36
    if fuechse <= 0 and maeuse > 0: fuechse = 1
    if maeuse <= 0: maeuse = 3; fuechse = 1

    # Daten für den Plot speichern
    historie_maeuse.append(maeuse)
    historie_fuechse.append(fuechse)

# --- VISUALISIERUNG ---
sns.set_theme(style="darkgrid")
plt.figure(figsize=(14, 5))
plt.plot(historie_maeuse, label='Mäuse (Beute) 🐭', color='#3498db', linewidth=2)
plt.plot(historie_fuechse, label='Füchse (Räuber) 🦊', color='#e67e22', linewidth=2)
plt.title('Euer erweitertes Ökosystem-Labor', fontsize=14, fontweight='bold')
plt.xlabel('Runden')
plt.ylabel('Anzahl Individuen')
plt.legend()
plt.show()
```
